```python
#заведём константы из условия и аналитической части
n1 = 2
n2 = 4
delta1 = 10**(-3)
delta2 = 10**(-6)
a=0.1
```

```python
from matplotlib import pyplot as plt
import numpy as np
from math import floor, pi, cos, sqrt
# импорт основных библиотек
```

```python
def func(x):
    return cos(x + pi / 4)
# наша функция

#теперь заведём функции для вычисления значений многочленов Тейлора порядка 0-n2 в точке x
def taylor0(x):
    return sqrt(2) / 2

def taylor1(x):
    return sqrt(2)/2-sqrt(2)/2*x

def taylor2(x):
    return sqrt(2)/2-sqrt(2)/2*x-sqrt(2)/4*x**2

def taylor3(x):
    return sqrt(2)/2-sqrt(2)/2*x-sqrt(2)/4*x**2+sqrt(2)/12*x**3

def taylor4(x):
    return sqrt(2)/2-sqrt(2)/2*x-sqrt(2)/4*x**2+sqrt(2)/12*x**3+sqrt(2)/48*x**4
```

```python
x=[] #массив абсцисс
y,yt0,yt1,yt2,yt3,yt4=[],[],[],[],[],[] #массив ординат для функции и многочленов
ibeg = -10.0
ilim = 10.0
```

```python
i = ibeg
while(i<=ilim):
    x.append(i)
    y.append(func(i))
    yt0.append(taylor0(i))
    yt1.append(taylor1(i))
    yt2.append(taylor2(i))
    yt3.append(taylor3(i))
    yt4.append(taylor4(i))
    i+=0.1
```

```python
# строим графики
fig, ax = plt.subplots(6, 1)
fig.set_figwidth(15)
fig.set_figheight(30)
fig.tight_layout()
plt.subplots_adjust(left=None, bottom=None, right=None, top=1.2, wspace=0.2, hspace=None)

# график последовательности
ax[0].set_xlim([ibeg-1, ilim+1])
ax[0].set_ylim([ibeg-1, ilim+1])
ax[0].set_title("Taylor 0")
ax[0].set_xticks(np.arange(ibeg-1,ilim+1,1.0))
ax[0].set_yticks(np.arange(ibeg-1,ilim+1,1.0))
ax[0].vlines(a,ibeg+3,ilim-3,linestyles='dotted')
text="f(a)={0} and Taylor_0(a)={1}".format(func(a),taylor0(a))
ax[0].annotate(text, xy=(-5.5, -9),size=14)

ax[1].set_xlim([ibeg-1, ilim+1])
ax[1].set_ylim([ibeg-1, ilim+1])
ax[1].set_title("Taylor 1")
ax[1].set_xticks(np.arange(ibeg-1,ilim+1,1.0))
ax[1].set_yticks(np.arange(ibeg-1,ilim+1,1.0))
ax[1].vlines(a,ibeg+3,ilim-3,linestyles='dotted')
text="f(a)={0} and Taylor_1(a)={1}".format(func(a),taylor1(a))
ax[1].annotate(text, xy=(-5.5, -9),size=14)

ax[2].set_xlim([ibeg-1, ilim+1])
ax[2].set_ylim([ibeg-1, ilim+1])
ax[2].set_title("Taylor 2")
ax[2].set_xticks(np.arange(ibeg-1,ilim+1,1.0))
ax[2].set_yticks(np.arange(ibeg-1,ilim+1,1.0))
ax[2].vlines(a,ibeg+3,ilim-3,linestyles='dotted')
text="f(a)={0} and Taylor_2(a)={1}".format(func(a),taylor2(a))
ax[2].annotate(text, xy=(-5.5, -8),size=14)
text1="|Taylor_2(a) - f(a)|={0} <= Delta1={1}".format(abs(taylor2(a)-func(a)),delta1)
ax[2].annotate(text1,xy=(-5.5,-9.5),size=14)

ax[3].set_xlim([ibeg-1, ilim+1])
ax[3].set_ylim([ibeg-1, ilim+1])
ax[3].set_title("Taylor 3")
ax[3].set_xticks(np.arange(ibeg-1,ilim+1,1.0))
ax[3].set_yticks(np.arange(ibeg-1,ilim+1,1.0))
ax[3].vlines(a,ibeg+3,ilim-3,linestyles='dotted')
text="f(a)={0} and Taylor_3(a)={1}".format(func(a),taylor3(a))
ax[3].annotate(text, xy=(-5.5, -9),size=14)

ax[4].set_xlim([ibeg-1, ilim+1])
ax[4].set_ylim([ibeg - 1, ilim+1])
ax[4].set_title("Taylor 4")
ax[4].set_xticks(np.arange(ibeg-1,ilim+1,1.0))
ax[4].set_yticks(np.arange(ibeg-1,ilim+1,1.0))
ax[4].vlines(a,ibeg+3,ilim-3,linestyles='dotted')
text="f(a)={0} and Taylor_4(a)={1}".format(func(a),taylor4(a))
ax[4].annotate(text, xy=(-5.5, -8),size=14)
text1="|Taylor_4(a) - f(a)|={0} <= Delta2={1}".format(abs(taylor4(a)-func(a)),delta2)
ax[4].annotate(text1,xy=(-5.5,-9.5),size=14)

ax[5].set_xlim([ibeg-1,ilim+1])
ax[5].set_ylim([ibeg-1,ilim+1])
ax[5].set_title('All in one')
ax[5].set_xticks(np.arange(ibeg-1,ilim+1,1.0))
ax[5].set_yticks(np.arange(ibeg-1,ilim+1,1.0))
ax[5].vlines(a,ibeg+3,ilim-3,linestyle='dotted')

# разные дизайны и настройки отображения
ax[0].plot(x,y,label='main func')
ax[0].plot(x,yt0,label='Taylor 0')
ax[0].legend()

ax[1].plot(x,y,label='main func')
ax[1].plot(x,yt1,label='Taylor 1')
ax[1].legend()

ax[2].plot(x,y,label='main func')
ax[2].plot(x,yt2,label='Taylor 2')
ax[2].legend()

ax[3].plot(x,y,label='main func')
ax[3].plot(x,yt3,label='Taylor 3')
ax[3].legend()

ax[4].plot(x,y,label='main func')
ax[4].plot(x,yt4,label='Taylor 4')
ax[4].legend()

ax[5].plot(x,y,label='main func')
ax[5].plot(x,yt0,label='Taylor 0')
ax[5].plot(x,yt1,label='Taylor 1')
ax[5].plot(x,yt2,label='Taylor 2')
ax[5].plot(x,yt3,label='Taylor 3')
ax[5].plot(x,yt4,label='Taylor 4')
ax[5].legend()

plt.show()
```
